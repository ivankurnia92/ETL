```python
import pandas as pd
from pandasql import sqldf
from sqlalchemy import create_engine
```

```python
# Konfigurasi database
db_uri = 'postgresql+psycopg2://postgres:ivankurnia92@localhost:5432/latihan'
```

```python
# Buat engine SQLAlchemy
engine = create_engine(db_uri)
```

```python
#read data csv files
df_sales_indonesia = pd.read_csv('../Project/sales_generated.csv')
df_sales_indonesia
```

```python
df_sales_indonesia.head(10)
```

```python
#transform process
#filter data name is not null
#and market_area is not null
#and location indonesia
#and number_of_sales more than 30
df_sales_idn_min_30 = sqldf(''' select * from df_sales_indonesia where
                     name is not null and market_area is not null and number_of_sales > 30  ''')
```

```python
df_sales_idn_min_30.head(10)
```

```python
count_data = df_sales_idn_min_30.shape[0]
count_data
```

```python
#load data after transformation to postgresql
TABLE_NAME = 'data_sales_idn_min_30'
df_sales_idn_min_30.to_sql(name=TABLE_NAME, con=engine,index=False)
print(f'Total Record has been inserted are {count_data} to table {TABLE_NAME} ')
```
